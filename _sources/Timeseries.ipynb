```python

# load the dataset
dataframe = pandas.read_csv('/content/drive/MyDrive/prosaindata/tugas/airline-passengers.csv',  usecols=[1], engine='python')
data = dataframe.values
data = data.astype('int')

# Membuat sample input dan output
look_back = 2
input_data, output_data = [], []
for i in range(len(data)-look_back-1):
    a = data[i:(i+look_back), 0]
    input_data.append(a)
    output_data.append(data[i + look_back, 0])

input_data = np.array(input_data)
output_data = np.array(output_data)

# Menampilkan sample input dan output
print("Sample input:")
print(input_data[:20])
print("\nSample output:")
print(output_data[:20])

df = pandas.DataFrame({'xt-2': input_data[:,0], 'xt-1': input_data[:,1], 'xt': output_data})
print(df)
df.to_csv('airlines_data.csv', index=False, columns=['xt-2', 'xt-1', 'xt'])
```

```python
import numpy as np

def split_sequence(sequence, n_steps, n_features):
    X, y = [], []
    for i in range(len(sequence)):
        # Mencari index akhir dari sequence yang ingin dipisahkan
        end_ix = i + n_steps
        # Memastikan index akhir tidak melebihi panjang sequence
        if end_ix > len(sequence)-1:
            break
        # Mendapatkan input dan output untuk sequence saat ini
        seq_x, seq_y = sequence[i:end_ix, :-1], sequence[end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    # Mengubah list menjadi array numpy
    X = np.array(X)
    y = np.array(y)
    return X, y

# Membaca dataset
dataset = np.loadtxt("/content/drive/MyDrive/prosaindata/tugas/airline-passengers.csv", delimiter=",", usecols=1, skiprows=1)

# Menambahkan kolom kedua ke dataset
dataset = np.column_stack((dataset, dataset))

# Menentukan jumlah time steps
n_steps = 2

# Memisahkan dataset menjadi input dan output menggunakan fungsi split_sequence
X, y = split_sequence(dataset, n_steps, n_features=2)

# Menampilkan sample input dan output
for i in range(len(X)):
  print(X[i], y[i])

df = pandas.DataFrame({'xt-2': X[:,0], 'xt-1': X[:,1], 'xt': y[]})
print(df)
```
