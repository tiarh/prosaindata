```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from gensim.models import Word2Vec

# Dataset
data = pd.read_csv('/content/drive/MyDrive/prosaindata/tugas/PTA.csv', sep=',')

# Preprocessing
data = data.dropna(subset=['Abstrak'])
corpus = data['Abstrak'].apply(lambda x: x.split())
model = Word2Vec(corpus, vector_size=100, window=5, min_count=1, workers=4)

# Transformasi corpus menjadi vektor
vectors = model.wv

# Transformasi corpus menjadi vektor
X = []
for i, sentence in enumerate(corpus):
    sentence_vector = np.zeros((100,))
    n_words = 0
    for word in sentence:
        if word in vectors.key_to_index:
            sentence_vector += vectors[word]
            n_words += 1
    if n_words > 0:
        sentence_vector /= n_words
    X.append(sentence_vector)

# Clustering
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# Labeling
labels = ['komputasi', 'pemrosesan bahasa alami', 'rekayasa perangkat lunak']
cluster_labels = [labels[i] for i in kmeans.labels_]

# Menambahkan kolom label pada dataset
data['label'] = cluster_labels

# Menyimpan hasil ke file csv
data.to_csv('hasil.csv', index=False)
```
