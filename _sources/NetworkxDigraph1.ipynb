```python
import networkx as nx
G = nx.DiGraph()
```

```python
# add edges to the graph
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 4)
G.add_edge(2, 5)
G.add_edge(3, 5)
G.add_edge(3, 6)
G.add_edge(4, 6)
G.add_edge(5, 7)
G.add_edge(6, 8)
G.add_edge(7, 9)
G.add_edge(8, 9)
G.add_edge(2, 9)
G.add_edge(5, 8)
G.add_edge(8, 7)
```

```python
nx.draw(G, with_labels=True)
```

```python
nx.pagerank(G, alpha=0.85)
```
