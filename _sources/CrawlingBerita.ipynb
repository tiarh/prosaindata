# Crawling Berita

```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import csv
import pandas as pd


hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv

def scrape_cbnc(hal):
    global hades
    a = 1

    with open('cnbc.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['headline', 'content'])

        for page in range(1, hal+1):
            url_cbnc = f"https://www.cnbcindonesia.com/search?query=covid+19&p={page}&kanal=&tipe=artikel&date="
            ge = req.get(url_cbnc,hades).text
            sop = bs(ge,'lxml')
            ul = sop.find('ul',class_='list media_rows middle thumb terbaru gtm_indeks_feed')
            li = ul.find_all('li')
            for x in li:
                link_art = x.find('article').find('a')['href']
                ge_ = req.get(link_art,hades).text
                sop_ = bs(ge_,'lxml')
                art = sop_.find('div',class_='lm_content mt10')
                if art is not None:
                    article = art.find('article')
                    if article is not None:
                        headline = article.find('h1').text.strip()
                        warp = article.find_all('div',class_='detail_text')
                        content = ""
                        for x in warp:
                            p_tags = x.find_all('p')
                            for p_tag in p_tags:
                                content += p_tag.text.strip() + " "
                        content = content.replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
                        print(f'done[{a}] > {headline[0:10]}')
                        a += 1
                        writer.writerow([headline, content])
                else:
                    print(f'done[{a}] > Article not found.')
                    a += 1
```

```python
scrape_cbnc(3)
```
