```python
import networkx as nx
G=nx.Graph()
```

```python
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(4,3),(4,6),(4,5),(5,6),(5,7),(5,8),(6,7),(6,8),(7,8),(7,9)])
```

```python
nx.draw(G, with_labels=True)
```

```python
nx.degree_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
nx.betweenness_centrality(G)
```
